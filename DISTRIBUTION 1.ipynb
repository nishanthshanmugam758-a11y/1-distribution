```python
from scipy.stats import norm

mu = 250
sigma = 25
x = 270
prob_greater = 1 - norm.cdf(x, mu, sigma)
print("probability students is taller than 270 cm:",round(prob_greater * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 250
sigma = 25

x= np.random.normal(mu, sigma, 10000)

plt.hist(x, bins=50, density=True, alpha=0.8)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, linewidth=2)

plt.title("Normal Distribution Histogram\nBell Shaped curve")
plt.xlabel("Height(cm)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
         
```

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

mean_time = 20
t = 7

lambda_value = 1 / mean_time
prob_within = expon.cdf(t, scale=1/lambda_value)

print("probability customer arrives within 7 minutes:",
      round(prob_within * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_time = 25

data = np.random.exponential(mean_time, 3000)

plt.hist(data, bins=60, density=True)

plt.title("Exponential Distribution - Histogram")
plt.xlabel("Time(minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
# Uniform Distribution range
a = 6      # minimum
b = 400    # maximum
# Interval to check
low = 40
high = 60
dist = uniform(loc=a, scale=b-a)
probability = dist.cdf(high) - dist.cdf(low)

print("Probability of getting number between 30 and 60:",
      round(probability * 180, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random uniform numbers between 1 and 100
data = np.random.uniform(6, 180, 10000)

# Plot histogram
plt.hist(data, bins=60, edgecolor='black')

plt.title("Uniform Distribution ")
plt.xlabel("Numbers (6 to 180)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
